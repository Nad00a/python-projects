```python
# determine which club has made the most transfers

clubs = spl_transfers['club_name'].value_counts()

print(clubs)
```

```python
plt.figure(figsize=(10, 6))
clubs.plot(kind='bar',color='orange')
plt.xlabel('Club')
plt.ylabel('Count')
plt.title('Number of Transfers by Club')
plt.xticks(rotation=90)
plt.show()
```

```python
# Fee column
#Fee column contains data with symbols, special characters so we need to do couple of steps
#1-investigate unique Data 

unique_values = spl_transfers['fee'].unique()
for value in unique_values:
    print(value)
```

```python
#2-perform some data cleaning methods to get rid of these values

# Replace values such as '?' and '-' with NaN
spl_transfers['fee'] = spl_transfers['fee'].replace(['?', '-'], np.nan)

# Remove values starting with 'end of loan'
spl_transfers['fee'] = spl_transfers['fee'].str.replace('^End of loan.*', '')

# Convert fee values to string
spl_transfers['fee'] = spl_transfers['fee'].astype(str)

# Remove '€' symbol
spl_transfers['fee'] = spl_transfers['fee'].str.replace('€', '')

# Replace 'm' with 'e6' and 'K' with 'e3'
spl_transfers['fee'] = spl_transfers['fee'].str.replace('m', 'e6').str.replace('K', 'e3')

# Replace 'free transfer' with '0'
spl_transfers['fee'] = spl_transfers['fee'].str.replace('free transfer', '0')

print(spl_transfers['fee'])
```

```python
club_names = clubs.index
count_values = clubs.values

# Create a bar plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(club_names, count_values)
plt.xlabel('Club')
plt.ylabel('Count')
plt.title('Number of Transfers by Club')
plt.xticks(rotation=90)
plt.show()
```

```python
#3- convert the type to numeric data type
spl_transfers['fee'] = pd.to_numeric(spl_transfers['fee'], errors='coerce')

# Check the data type of 'fee' column after conversion
print(spl_transfers['fee'].dtype)
```

```python
#top 20 deals in terms of price

top20 = spl_transfers.nlargest(20, 'fee')

# Extract the 'club_name', 'player_name' and 'fee' columns from the top 20 expensive deals
output = top20[['club_name','player_name', 'fee']]

# Print the output
print(output)
```

```python
fees = output['fee']
player_names = output['player_name']

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(fees, labels=player_names, autopct='%1.1f%%')
plt.title('Top 20 Expensive Deals')
plt.show()
```
